```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
data = {
    "Day": np.arange(1, 31),
    "Call_Volume": [120, 135, 150, 160, 155, 170, 180, 175, 165, 160,
                    158, 162, 170, 180, 190, 200, 195, 185, 180, 175,
                    170, 165, 168, 172, 178, 185, 190, 195, 200, 210],
    "Sentiment_Score": [-0.2, -0.1, 0.1, 0.2, 0.15, 0.05, -0.05, 0.1,
                        0.2, 0.3, 0.25, 0.2, 0.15, 0.1, 0.05, -0.1,
                        -0.15, 0.05, 0.1, 0.15, 0.2, 0.25, 0.3,
                        0.35, 0.4, 0.45, 0.5, 0.55, 0.6, 0.65]
}
df = pd.DataFrame(data)
print("Sample Call Center Data")
print(df.head())
plt.figure()
plt.plot(df["Day"], df["Call_Volume"], marker='o')
plt.xlabel("Day")
plt.ylabel("Number of Calls")
plt.title("Daily Call Volume Trend")
plt.grid()
plt.show()
plt.figure()
plt.plot(df["Day"], df["Sentiment_Score"], marker='o')
plt.xlabel("Day")
plt.ylabel("Sentiment Score")
plt.title("Customer Sentiment Trend")
plt.grid()
plt.show()
X = df[["Day"]]
y = df["Call_Volume"]
model = LinearRegression()
model.fit(X, y)
future_days = pd.DataFrame({"Day": np.arange(31, 38)})
future_predictions = model.predict(future_days)
future_df = pd.DataFrame({
    "Day": future_days["Day"],
    "Predicted_Call_Volume": future_predictions.astype(int)
})
print("\nPredicted Call Volume for Next 7 Days")
print(future_df)
plt.figure()
plt.plot(df["Day"], df["Call_Volume"], label="Actual Calls")
plt.plot(future_df["Day"], future_df["Predicted_Call_Volume"],
         linestyle='--', marker='o', label="Predicted Calls")
plt.xlabel("Day")
plt.ylabel("Number of Calls")
plt.title("Call Volume Forecasting")
plt.legend()
plt.grid()
plt.show()
avg_sentiment = df["Sentiment_Score"].mean()
print("\nAverage Customer Sentiment Score:", round(avg_sentiment, 2))
if avg_sentiment > 0:
    print("Overall customer sentiment is POSITIVE")
else:
    print("Overall customer sentiment is NEGATIVE")
```
